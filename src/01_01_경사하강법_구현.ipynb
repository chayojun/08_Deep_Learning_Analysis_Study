# 경사하강법 구현

```python
import numpy as np  # 숫자 계산을 도와주는 도구(라이브리)
import matplotlib.pyplot as plt  # 그래프를 그려주는 도구
```

```python
X = np.random.rand(100)  # 0과 1 사이의 랜덤(무작위) 숫자 100개를 만들어서 X에 넣는 코드
y = X * 0.7  # 이건 '정답' 선생님(데이터 만드는 사람)이 정해둔 규칙

plt.figure(figsize=(8, 6))  # 그래프를 그릴 도화지(캔버스)를 만들어요. 크기를 가로8, 세로6
plt.scatter(X, y)  # X와 y를 점으로 찍어요. X는 가로, y는 세로로 표시

# X = '시험문제'
# y = '정답'
```

```python
# -1과 1 사이에서 랜덤(무작위)으로 하나의 실수(소수)를 뽑아서 w에 넣는 것

w = np.random.uniform(-1,1)
w
```

```python
# -1과 1 사이에서 랜덤(무작위)으로 하나의 실수(소수)를 뽑아서 B에 넣는 것

b = np.random.uniform(-1,1)
b
```

```python
# 넘파이 배열은 더 빠른 계산, 벡터 연산, 수학 함수 사용이 가능
# 평균 = (1 + 2 + 3 + 4 + 5) ÷ 5 = 15 ÷ 5 = 3.0

np.array([1,2,3,4,5]).mean()
```

```python

lr = 0.5  # 학습률(한 번에 얼마만큼 업데이트할지)


for epoch in range(300):  # epoch 반복문: 전체 데이터를 한 번씩 보고 배우는 동작을 300번 반복
    y_pred = w*X + b  # 예측값 계산: 현재의 w(기울기)와 b(절편)를 사용해서 모든 X에 대한 예측 y를 만듬

    error = np.abs(y_pred - y).mean()  # 손실(오차)을 계산: 각 예측과 실제 정답의 차이(|y_pred - y|)을 평균

    w_grad = lr * ((y_pred-y)*X).mean()  
    b_grad = lr * (y_pred-y).mean()

    w = w - w_grad  # w를 w_grad만큼 빼요. (오차를 줄이는 방향으로 한 걸음 이동)
    b = b - b_grad   # b도 같은 방식으로 업데이트

    print('error : ',error)   # 현재 오차를 콘솔에 찍어줘서 학습이 잘 되고 있는지 확인

    if epoch%10 == 0:  # epoch을 10번마다(0,10,20...)마다 현재 상태의 그래프를 그림
        y_pred = w * X + b
        y_pred = w*X + b
        plt.figure(figsize=(8,6))
        plt.scatter(X,y)  # 실제 데이터
        plt.scatter(X,y_pred)  # 예측 데이터
        plt.show()

    if error < 0.001:  # 오차가 충분히 작아지면 중단
        break

```
